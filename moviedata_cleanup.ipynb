```python
import numpy as np 
import pandas as pd
filename='F://data analyst//project//P3//movies.csv'
movies_df=pd.read_csv(filename)
movies_df.info()
```

```python
#select the variables I need during visualization:
#popularity,budget,revenue,original_title,keywords,genres,production-companies,release_year,vote_count,vote_average
#then create a new DataFrame
newmovies_df=movies_df[['id','popularity','budget_adj','revenue_adj','original_title','cast','keywords','genres','production_companies','vote_count','vote_average','release_year']]
newmovies_df.info()
```

```python
#Work with missing data
#Drop rows with missing data
newmovies_df=newmovies_df.dropna()
newmovies_df.info()
```

```python
split_production_companies=newmovies_df['production_companies'].str.split('|',expand=True)
split_production_companies['id']=newmovies_df['id']
```

```python
pc1=split_production_companies.copy()
def P_or_U(x):
    if x=='Paramount Pictures': 
        return 'P'
    elif x=='Universal Pictures':
        return 'U'
    else:
        return ''
for a in range(5):
    pc1[a]=pc1[a].apply(P_or_U)
pc1['P_or_U']=pc1[0]+pc1[1]+pc1[2]+pc1[3]+pc1[4]
pc1=pc1.replace('','else')
```

```python
newmovies_df['P_or_U']=pc1['P_or_U']
newmovies_df.head()
```

```python
split_keywords=newmovies_df['keywords'].str.split('|',expand=True)
split_keywords['id']=newmovies_df['id']
```

```python
for a in range(5):
    split_keywords[a]=split_keywords[a]=='based on novel'
split_keywords['novel']=split_keywords[0] | split_keywords[1] | split_keywords[2] | split_keywords[3] | split_keywords[4]
newmovies_df['novel_or_not']=split_keywords['novel']
newmovies_df.head(10)
```

```python
newmovies_df.to_csv('F://data analyst//project//P3//newmovies.csv',index=False)
```

```python
split_genres=newmovies_df['genres'].str.split('|',expand=True)
split_genres['id']=newmovies_df['id']
merged1=newmovies_df.merge(split_genres)
melted1=pd.melt(
    merged1,id_vars=['id','original_title','release_year','budget_adj','revenue_adj','popularity','vote_count','vote_average','P_or_U','novel_or_not'],
    value_vars=[0,1,2,3,4],value_name='genres')
melted1=melted1.drop('variable',axis=1).dropna()
melted1.head()
```

```python
melted1.to_csv('F://data analyst//project//P3//movies_cleanup.csv',index=False)
```
